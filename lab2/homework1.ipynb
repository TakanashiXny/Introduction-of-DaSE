1. 判断BMI指数

```python
weight=float(input('请输入您的体重(单位: 千克): '))
height=float(input('请输入您的身高(单位: 米): '))
bmi=weight/(height**2)
print(bmi)
if bmi<18.5:
    print("轻体重")
elif 18.5 <= bmi < 24:
    print("正常")
elif 24 <= bmi < 28:
    print("超重")
else:
    print("肥胖")
```

```python
for i in range(1,10):
    for j in range(1,i+1):
        print(f"{j}*{i}={j*i}",end="\t")
    print("")
```
