```python
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv("college_entrance.csv")
data
```

```python
data = data.drop([31, 32])
data
```

```python
avg = data.copy()
```

```python
avg.head()
```

```python
lst = avg.columns.tolist()[2:]
lst
```

```python
for i in lst:
    avg[i] = avg[i] / avg['高考人数']
avg
```

```python
Beijing_University = avg.sort_values(by=['北京大学'])
plt.figure(figsize=(10, 10))
x = list(Beijing_University['地区'])
y = list(Beijing_University['北京大学'])
norm = plt.Normalize(min(y), max(y))
normy = norm(y)
map_vir = cm.get_cmap(name='viridis')
color = map_vir(normy)
plt.barh(x, y, color=color)
plt.ylabel('城市',fontdict={'size':20})
plt.xlabel('录取比例', fontdict={'size':15})
plt.title('北京大学各省录取比例 ', fontdict={'size':16})
plt.show()
```
