```python
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv("Education.csv")
data
```

此数据集取自《2020年第七次人口普查主要数据》p24

```python
lst = ['大专及以上', '高中', '初中', '小学']
for i in lst:
    data[i] = data[i] / 1000
data
```

大专及以上

```python
data1 = data.sort_values(by=['大专及以上'])
data1
```

```python
lst = list(data1['大专及以上'])
lst1 = list(data1['地区'])
norm = plt.Normalize(min(lst), max(lst))
norm_y = norm(lst)
plt.figure(figsize=(10, 10))
map_vir = cm.get_cmap(name='viridis')
color = map_vir(norm_y)
plt.barh(lst1, lst, color=color)
plt.ylabel('城市',fontdict={'size':20})
plt.xlabel('大专及以上比例(%)', fontdict={'size':15})
plt.title('大专及以上比例', fontdict={'size':16})
plt.show()
```
